```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = len(purchase_data["Item ID"].unique())
total_revenue = purchase_data["Price"].sum()
no_of_purchases = len(purchase_data)
average_price = round(total_revenue / no_of_purchases, 2) 

summary_df = pd.DataFrame({
    "Number of Unique Items": [unique_items],
    "Average Price": [average_price],
    "Number of Purchases": [no_of_purchases],
    "Total Revenue": [total_revenue]
})
summary_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
players = purchase_data.groupby("Gender")
players_count = players["SN"].nunique()
player_percent = round(players["SN"].nunique() / total_players * 100, 2)

gender_df = pd.merge(players_count, player_percent, on="Gender")
gender_df = gender_df.rename(columns={'SN_x': 'Total Count', 'SN_y': 'Percentage of Players'})
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_grouped = purchase_data.groupby("Gender")
purchase_count = gender_grouped["Purchase ID"].count()
total_purchases = gender_grouped["Price"].sum()
average_purchase_price = round(total_purchases / purchase_count, 2) 
average_purchase_per_person = round(total_purchases / players_count, 2)
                  
purchasing_analysis_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": average_purchase_price,
    "Total Purchase Value": total_purchases,
   "Avg Total Purchase per Person": average_purchase_per_person
})
purchasing_analysis_df.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]
group_labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'] 

print(purchase_data["Age"].min())
print(purchase_data["Age"].max())

bin_df = purchase_data
bin_df["Age group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
bin_df.head()

bin_df = bin_df.groupby("Age group")
bin_df.head()
age_count = bin_df["SN"].nunique()
total_unique_counts = round(age_count / 576 * 100, 2)

age_group = pd.merge(age_count, total_unique_counts, on="Age group", how='left')
age_group = age_group.rename(columns={'SN_x': 'Total Count', 'SN_y': 'Percentage of Players'})
age_group
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_df = bin_df
purchase_count = purchase_df["Purchase ID"].count()
avg_purchase_price = round(purchase_df["Price"].mean(), 2)
total_purchase = purchase_df["Price"].sum()
average_purchase_per_person = round(total_purchase / age_group["Total Count"], 2) 

purchase_age_analysis = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase_price,
    "Total Purchase Value": total_purchase,
    "Avg Total Purchase per Person": average_purchase_per_person
})
purchase_age_analysis.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_df = purchase_data
spender_df = spender_df.groupby("SN")
purchase_count = spender_df["SN"].count()
purchase_price = round(spender_df["Price"].mean(), 2)
total_purchases = spender_df["Price"].sum()

top_spender_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": purchase_price,
    "Total Purchase Value": total_purchases
})
top_spender_df.sort_values("Total Purchase Value",ascending=False).head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_df = purchase_data
popular_df = popular_df.groupby(["Item ID", "Item Name", "Price"])
purchase_count = popular_df["Item ID"].count()
total_purchase = popular_df["Price"].sum()

most_popular_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Total Purchase Value": total_purchase
})
most_popular_df.sort_values("Purchase Count", ascending=False).head()
```
